```python
#Fetching our price data for the FAANG stocks 
tickers = ['FB', 'AMZN', 'AAPL', 'NFLX', 'GOOG']
price_data = pdr.data.get_data_yahoo(tickers,
                           start = '2020-01-01',
                           end = '2020-05-11')['Adj Close']
```

```python
#Here we are calculating the logarithmic return of the price data of FAANG - the Shift method shifts the values by 1 column
log_ret = np.log(price_data/price_data.shift(1))
```

```python
#Creating the covariance matrix, we multiply by 252 to represent the number of trading days in a year
cov_mat = log_ret.cov() * 252
print(cov_mat)
```

```python
# Simulating 10000 portfolios
num_port = 10000
# Creating an empty array to store portfolio weights
all_wts = np.zeros((num_port, len(price_data.columns)))
# Creating an empty array to store portfolio returns
port_returns = np.zeros((num_port))
# Creating an empty array to store portfolio risks
port_risk = np.zeros((num_port))
# Creating an empty array to store portfolio sharpe ratio
sharpe_ratio = np.zeros((num_port))
```

```python
#Setting up the for loop to simulate the returns and risks of 10,000.
for i in range(num_port):
  wts = np.random.uniform(size = len(price_data.columns))
  wts = wts/np.sum(wts)

 #The colon in this array simply means that the array goes front i to the end of the array
  all_wts[i,:] = wts
  
  #We then bring together the returns of our portfolio and save this in our array 
  port_ret = np.sum(log_ret.mean() * wts)
  port_ret = (port_ret + 1) ** 252 - 1
  
  
  port_returns[i] = port_ret
  
  
  #Saving the standard deviation of the FAANG stock in yet another array
  port_sd = np.sqrt(np.dot(wts.T, np.dot(cov_mat, wts)))
  
  port_risk[i] = port_sd
  
  #Saving out Sharpe ratio in its own array, we also assume a risk free rate of 0%
  sr = port_ret / port_sd
  sharpe_ratio[i] = sr
```

```python
#Finding the minimum variance weights of each of our FAANG stocks
names = price_data.columns
min_var = all_wts[port_risk.argmin()]
print(min_var)
```

```python
#Finding the maximum Sharpe Ratio weights of each of our FAANG stocks
max_sr = all_wts[sharpe_ratio.argmax()]
print(max_sr)
```

```python
#This is our maximum Sharpe Ratio, a Sharpe ratio of 2.3 is considered good
print(sharpe_ratio.max())
```

```python
#This is our minimum possible risk portfolio. 0.44 is the Beta of the minimum risk portfolio which means that the stock moves by about 44% of the index i.e. it is relatively safe 
print(port_risk.min())
```

```python
#Plotting our minimum variance portfolio weights for visualization
min_var = pd.Series(min_var, index=names)
min_var = min_var.sort_values()
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.set_xlabel('Asset')
ax1.set_ylabel("Weights")
ax1.set_title("Minimum Variance Portfolio weights")
min_var.plot(kind = 'bar')
plt.show();
```

```python
#The Tangency portfolio weights is another way to say that we are looking for the weights that would maximize our Sharpe Ratio
max_sr = pd.Series(max_sr, index=names)
max_sr = max_sr.sort_values()
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.set_xlabel('Asset')
ax1.set_ylabel("Weights")
ax1.set_title("Tangency Portfolio weights")
max_sr.plot(kind = 'bar')
plt.show();
```

```python
#Putting this all together, we put together the Efficient Frontier, with the y axis representing our returns, and the x axis representing risk
#The most optimal portfolio is on the upper left, which represent the portfolio with the lowest risk and the highest return
#The least optimal portfolio is on the bottom right, with a beta of 0.58 and a return of only 20%
#Most portfolios typically lie in the middle of the efficient frontier
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.set_xlabel('Risk')
ax1.set_ylabel("Returns")
ax1.set_title("Portfolio optimization and Efficient Frontier")
plt.scatter(port_risk, port_returns)
plt.show();
```
